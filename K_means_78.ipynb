```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
#load iris dataset
iris=load_iris()
X=iris.data
y=iris.target
```

```python
iris
```

```python
df=pd.DataFrame(X)
df
```

```python
#implement elbow method
wcss=[] #Within-Cluster Sum of Squares

for i in range(1,11):
  kmeans =KMeans(n_clusters=i,n_init=10,random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
#plot the elbow method
plt.plot(range(1,11),wcss,marker="*")
plt.title("Elbow Method")
```

```python
k_optimal=3
kmeans=KMeans(n_clusters=k_optimal,init='k-means++',max_iter=300,n_init=10,random_state=0)
kmeans.fit(X)
```

```python
#get centroids
centroids=kmeans.cluster_centers_
labels=kmeans.labels_
```

```python
#visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='o')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Clustered Classification with Centroids')
plt.show()
```
